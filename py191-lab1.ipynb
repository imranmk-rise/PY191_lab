## 3. Variables & Data Types
Python is **dynamically typed** – variable types are inferred at runtime.

| Type | Example | Literal |
|------|---------|---------|
| `int` | whole numbers | `x = 42` |
| `float` | decimal numbers | `pi = 3.14159` |
| `str` | text | `name = "Ada"` |
| `bool` | truth values | `is_valid = True` |
| `list` | ordered, mutable collection | `scores = [95, 88, 72]` |
| `dict` | key‑value pairs | `user = {'id': 1, 'name': 'Sam'}` |

Run the cell below to see the types in action.

```python
x = 42
pi = 3.14159
name = "Ada"
is_valid = True
scores = [95, 88, 72]
user = {'id': 1, 'name': 'Sam'}

print(type(x), type(pi), type(name), type(is_valid), type(scores), type(user))
```

## Activity 4 – Factorial Function
Implement `factorial(n)` that returns the factorial of n (`factorial(0)=1`, `factorial(1)=1`, `factorial(2)=2`, etc.). Test that `factorial(5)` is 120.

```python
# Your code here


def factorial(n: int):
    x = 1
    for i in range(1, n+1):
        x *= i
    
    return x

print(factorial(5))
```

```python
# prime checker function

# 1. is the item odd?
# --> check if the item is a multiple of 3 (i%3==0) or 5
# --> if the item is a multiple of 3 or 5, is_prime(n) returns True
# --> if not, create a list of all numbers from 7 to n-1, excluding multiples of 3 and 5 and all even numbers, and check if n is divisible by those numbers
# 2. if the item is even, is_prime(n) returns False

def is_prime(n: int):
    if n == 1 or n == 2 or n == 3 or n == 5:
        return True
    if n%2==0:
        return False
    if n%3==0 or n%5==0:
        return False
    
    arr = [i for i in range(7, int(n**0.5)+1) if i%3!=0 and i%5!=0 and i%2!=0]
    for i in arr:
        if n%i==0:
            return False
    return True

is_prime(n=25)


# are 2 3 and 5 prime?

```
